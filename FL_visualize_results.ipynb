```python
from logging import WARNING, INFO
from typing import Callable, Dict, List, Optional, Tuple, Union
import numpy as np
import yaml
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
run_name = '16_32'
save_path = 'plots/'
file_path = 'outputs/FL/' #NEED TO READ ALL FILES
num_epochs = 102
```

```python
# Read file
file = open(file_path+run_name+'.out', mode = 'r', encoding = 'utf-8-sig')
lines = file.readlines()
file.close()
```

```python
tmp1 = []
for line in lines:
    tmp1.append((line.split(' ')))
```

```python
loss = []
acc = []
for i in range(num_epochs):
    loss.append(float(tmp1[1][2*(i+1)][:-1]))
    acc.append(float(tmp1[4][2*(i+1)][:-1]))
    
```

```python
plt.plot(np.arange(num_epochs), acc)
plt.ylim(0, 1)
#x = np.arange(0, num_agents, 1)
#plt.xticks(x)
#plt.title("Tplgy 0", fontsize=9)
plt.xlabel('Comm Round')
plt.ylabel('Acc')
plt.grid()

#ax = plt.gca()
#box = ax.get_position()
#ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
#ax.legend(['Tplgy 0', 'Tplgy 1', 'Tplgy 2', 'Tplgy 3', 'Tplgy 4'], loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig(save_path+run_name+'_FL_summarized_acc.pdf')
```
